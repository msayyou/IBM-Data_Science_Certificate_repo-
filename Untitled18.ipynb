```python
!pip install pymc arviz numpy numpy-financial

import numpy as np
import numpy_financial as npf
import pymc as pm
import arviz as az

# -----------------------------
# 1) Observed data (synthetic)
# -----------------------------
rng = np.random.default_rng(42)

rooms = 120

# 12 months of available nights (approx; you can use exact calendar days)
days_in_month = np.array([31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31])
available_nights = rooms * days_in_month

# Observed sold room-nights (integer counts). This is what you typically have from ops reports.
sold_nights = np.array([2700, 2550, 2780, 2480, 2900, 2650, 2950, 3000, 2600, 2850, 2700, 2750])

# Observed ADR (e.g., from STR / PMS)
adr_obs = np.array([148, 152, 155, 149, 160, 158, 165, 168, 156, 162, 159, 161], dtype=float)

# -----------------------------
# 2) Bayesian model
# -----------------------------
with pm.Model() as model:
    # --- Priors informed by "expert belief / comps" ---
    # Long-run stabilized occupancy (proportion)
    occ = pm.Beta("occ", alpha=80, beta=20)  # mean ~ 0.80

    # ADR level and month-to-month dispersion
    adr_mean = pm.Normal("adr_mean", mu=155, sigma=20)
    adr_sigma = pm.HalfNormal("adr_sigma", sigma=10)

    # Expense ratio (Operating Expenses / Total Revenue)
    # (Pick parameters to reflect your beliefs; mean here ~0.65)
    expense_ratio = pm.Beta("expense_ratio", alpha=65, beta=35)

    # Annual RevPAR growth (can be negative; truncated for realism)
    revpar_growth = pm.TruncatedNormal("revpar_growth", mu=0.02, sigma=0.015, lower=-0.05, upper=0.10)

    # Exit cap rate (positive, typically 5%–12% range)
    exit_cap = pm.TruncatedNormal("exit_cap", mu=0.08, sigma=0.01, lower=0.04, upper=0.12)

    # Recurring annual CAPEX (positive, skewed)
    # Gamma in PyMC uses alpha (shape) and beta (rate). Mean = alpha/beta.
    capex_annual = pm.Gamma("capex_annual", alpha=2.0, beta=2.0 / 120_000.0)  # mean ~ 120k

    # --- Likelihoods: how the data is generated ---
    # Occupancy evidence via sold nights out of available nights
    pm.Binomial("sold_nights", n=available_nights, p=occ, observed=sold_nights)

    # ADR evidence via noisy monthly observations around adr_mean
    pm.Normal("adr", mu=adr_mean, sigma=adr_sigma, observed=adr_obs)

    # Sample from posterior
    trace = pm.sample(
        draws=3000,
        tune=1500,
        target_accept=0.9,
        random_seed=42
    )

# Quick posterior summary (parameters)
print(az.summary(trace, var_names=["occ", "adr_mean", "expense_ratio", "revpar_growth", "exit_cap", "capex_annual"],
                 hdi_prob=0.95))

# -----------------------------
# 3) Convert posterior draws into an IRR distribution
# -----------------------------
posterior = trace.posterior.stack(sample=("chain", "draw"))

occ_s = posterior["occ"].values
adr_s = posterior["adr_mean"].values
exp_s = posterior["expense_ratio"].values
g_s   = posterior["revpar_growth"].values
cap_s = posterior["exit_cap"].values
cx_s  = posterior["capex_annual"].values

n_sims = occ_s.shape[0]

# Investment / underwriting assumptions (edit these!)
purchase_price = 25_000_000
acq_cost_pct = 0.02          # 2% acquisition costs
sell_cost_pct = 0.02         # 2% disposition costs
other_revenue_factor = 1.10  # total revenue = rooms revenue * 1.10 (adds F&B/other)
hold_years = 5

initial_outlay = -(purchase_price * (1 + acq_cost_pct))

# Compute cash flows for each posterior draw
irrs = np.empty(n_sims, dtype=float)
irrs[:] = np.nan

for i in range(n_sims):
    occ_i = occ_s[i]
    adr_i = adr_s[i]
    exp_i = exp_s[i]
    g_i   = g_s[i]
    cap_i = cap_s[i]
    cx_i  = cx_s[i]

    # Year-1 rooms revenue baseline from posterior occ & ADR
    revpar1 = occ_i * adr_i
    rooms_rev1 = revpar1 * rooms * 365.0
    total_rev1 = rooms_rev1 * other_revenue_factor

    nois = []
    for t in range(1, hold_years + 1):
        total_rev_t = total_rev1 * (1.0 + g_i) ** (t - 1)
        opex_t = total_rev_t * exp_i
        noi_t = total_rev_t - opex_t - cx_i
        nois.append(noi_t)

    noi_5 = nois[-1]
    exit_value = noi_5 / cap_i
    net_sale_proceeds = exit_value * (1 - sell_cost_pct)

    cashflows = [initial_outlay] + nois[:-1] + [noi_5 + net_sale_proceeds]

    irr = npf.irr(cashflows)
    irrs[i] = irr

# Remove any invalid IRRs (can happen if cashflow sign changes are odd)
irrs = irrs[np.isfinite(irrs)]

# -----------------------------
# 4) Decision outputs
# -----------------------------
target_irr = 0.12

print("\nPosterior IRR results (annual):")
print(f"Mean IRR:   {irrs.mean():.3%}")
print(f"Median IRR: {np.median(irrs):.3%}")
print(f"5% / 95%:   {np.quantile(irrs, 0.05):.3%} / {np.quantile(irrs, 0.95):.3%}")
print(f"P(IRR >= {target_irr:.0%}): {(irrs >= target_irr).mean():.1%}")

# Optional: plot posteriors
# az.plot_posterior(trace, var_names=["occ", "adr_mean", "expense_ratio", "revpar_growth", "exit_cap", "capex_annual"])
# az.plot_kde(irrs)
```

```python
import pymc as pm
import arviz as az
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------------------------
# SCENARIO: We are evaluating a 125-room limited service hotel in Austin, TX
# INVESTMENT RULE: We only approve deals with >70% chance of hitting 12% IRR
# ------------------------------------------------------------------------------

# --------------------------
# Step 1: Input our observed data
# Actual operating results from the last 6 months
# --------------------------
observed_monthly_occupancy = np.array([0.72, 0.77, 0.75, 0.79, 0.73, 0.76])
observed_monthly_adr = np.array([138, 149, 143, 154, 140, 148])

n_rooms = 125
holding_period = 5
purchase_price = 15_000_000
hurdle_rate = 0.12

# --------------------------
# Step 2: Define our priors
# This is where we formally encode all expert knowledge & market research
# Exactly as described in your framework
# --------------------------
with pm.Model() as hotel_investment_model:

    # --- Priors: Our beliefs before seeing new operating data ---
    # Occupancy is a proportion 0-1, so we use a Beta distribution
    # Rationale: Comps in this submarket average 72% occupancy, 95% range 64% to 79%
    occupancy = pm.Beta("occupancy", alpha=72, beta=28)

    # Average Daily Rate
    # Rationale: Brand data shows this asset class should do $145 ADR +/- $15
    adr = pm.Normal("adr", mu=145, sigma=15)

    # Net Operating Margin
    # Rationale: Limited service hotels consistently run 40-60% margin
    margin = pm.Beta("margin", alpha=50, beta=50)

    # Exit cap rate at end of 5 year hold
    cap_rate = pm.Normal("cap_rate", mu=0.10, sigma=0.01)

    # --------------------------
    # Step 3: Likelihood: Update our priors with new data
    # THIS IS BAYES THEOREM IN ACTION
    # We combine our prior beliefs with real observed performance
    # --------------------------
    pm.Binomial(
        "observed_occupancy",
        n=n_rooms,
        p=occupancy,
        observed=observed_monthly_occupancy * n_rooms
    )

    pm.Normal(
        "observed_adr",
        mu=adr,
        sigma=10,
        observed=observed_monthly_adr
    )

    # --------------------------
    # Step 4: FULL PROBABILISTIC DCF MODEL
    # This is the magic part no one tells you about:
    # We embed our entire financial model directly inside the Bayesian inference
    # --------------------------
    revpar = pm.Deterministic("revpar", occupancy * adr)
    annual_noi = pm.Deterministic("annual_noi", revpar * n_rooms * 365 * margin)
    exit_value = pm.Deterministic("exit_value", annual_noi / cap_rate)

    # Calculate NPV for this scenario
    # NPV > 0 is exactly equivalent to IRR > our 12% hurdle
    npv = pm.Deterministic(
        "npv",
        np.sum([annual_noi / (1 + hurdle_rate)**t for t in range(1,6)])
        + exit_value / (1 + hurdle_rate)**5
        - purchase_price
    )

    # Directly calculate the probability we hit our target
    prob_hit_hurdle = pm.Deterministic("prob_hit_hurdle", pm.math.switch(npv > 0, 1, 0))

    # --------------------------
    # Step 5: Run MCMC sampling
    # --------------------------
    trace = pm.sample(
        draws=5000,
        tune=2000,
        cores=4,
        random_seed=42
    )

# High level summary of all outputs
az.summary(trace, var_names=["occupancy", "adr", "revpar", "prob_hit_hurdle"], round_to=2)

# Plot the single most important chart for your investment committee
az.plot_posterior(trace, var_names=["npv"], ref_val=0, figsize=(12,5))
plt.title("Distribution of Net Present Value | Hurdle Rate 12%")
plt.show()
```
